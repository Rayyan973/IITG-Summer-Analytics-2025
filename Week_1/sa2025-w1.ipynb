### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
df = df.dropna(subset=['horsepower', 'weight'])
#this part is to remove NaN values in the horsepower and weight columns

filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

most_common = filtered_df['origin'].mode()[0]
print("most common origin = ", most_common)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
#first find the median
median = df['mpg'].median()

filtered_df = df[df['mpg'] > median].copy()
filtered_df['hp:weight'] = (filtered_df['horsepower']/filtered_df['weight'])
best_car = filtered_df.loc[filtered_df['hp:weight'].idxmax()]
print('best car = ', best_car['name'])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
grouped = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

sns.set(style='whitegrid')

plt.figure(figsize=(8, 4))
for origin in grouped['origin'].unique():
    data = grouped[grouped['origin'] == origin]
    plt.plot(data['model_year'], data['mpg'], marker='o', label=origin)

plt.title('average mpg over the years')
plt.xlabel('model year')
plt.ylabel('avg mpg')
plt.legend(title='origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
sns.set(style='whitegrid')

plt.figure(figsize=(8, 4))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    palette='Set1',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7,
    edgecolor='gray'
)

plt.title('Horsepower vs Weight of Cars by Origin (Size = MPG)', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
